### Observação:
Para se plotar mais de um gráfico, precisamos definir o conceito de figura.
- Numa figura é possível se definir diversos subplots.
   - criação da figura  --> figure(num_figura)
   - criação do subplot --> subplot(num_linha, num_coluna, num_grafico)

````python
# Exemplo de subplots
# Plotando mais de um plot numa mesma figura
from matplotlib.pylab import *
figure() # criação da figura

subplot(2, 1, 1)

t = linspace(0, 3, 51)
y1 = f1(t)
y2 = f2(t)
t3 = t[::4]
y3 = f2(t3)

#  ------ definição do subplot 1   -----------
subplot(2, 1, 1)
plot(t, y1, "r-", t, y2, "bo")
xlabel("t")
ylabel("y1")
axis([t[0], t[-1], min(y2)-0.05, max(y2)+0.5])
legend([r"$t^2e^{-t^2}$", r"$t^4e^{-t^2}$"])
title("Título do sub_plot1")


#  ------ definição do subplot 2   -----------
subplot(2, 2, 4)
plot(t, y1, "b-", t3, y3, "ys")
xlabel("t")
ylabel("y2")
axis([0, 4, -0.2, 0.6])
legend([r"$t^2e^{-t^2}$", r"$t^4e^{-t^2}$"])

tight_layout()
show()
````

### 📊 Usando `fig.add_axes()` para Criar Eixos no Matplotlib

O método **`fig.add_axes()`** é uma maneira flexível e poderosa de criar e posicionar sistemas de coordenadas dentro de uma figura no Matplotlib. Utilizamos este método quando desejamos um controle preciso sobre a posição e o tamanho dos gráficos.

#### 💡 Contexto:
O método **`add_axes()`** é aplicado a um objeto do tipo **`figure`** e permite criar **eixos** (ou sistemas de coordenadas) manualmente dentro dessa figura. Isso é útil quando precisamos adicionar gráficos sobrepostos, gráficos secundários (insets) ou controlar a posição de múltiplos gráficos na mesma tela.

---

#### 📝 Sintaxe:
```python
fig.add_axes([left, bottom, width, height])
```

#### 🗺️ **Parâmetros:**
O método **`add_axes()`** recebe uma lista com quatro valores normalizados (entre 0 e 1), que representam a proporção em relação ao tamanho da figura:

- **left:** Distância da borda esquerda da figura até o início do eixo.  
- **bottom:** Distância da borda inferior da figura até o início do eixo.  
- **width:** Largura do eixo em relação à largura total da figura.  
- **height:** Altura do eixo em relação à altura total da figura.  

> 💡 Os valores são normalizados para o intervalo [0, 1], onde:
> - `0` representa a borda inicial (esquerda ou inferior).  
> - `1` representa a borda final (direita ou superior).  

---

#### 📌 **Dica:**
Utilize **`fig.add_axes()`** quando precisar de um controle absoluto sobre o posicionamento dos gráficos. Para layouts automáticos ou ajustáveis, prefira métodos como **`plt.subplots()`** ou **`gridspec`**.

````python
# Exemplo de gráficos superpostos

from pylab import *

x = linspace(0, 5, 10)
y = x ** 2

fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # Eixo principal na figura
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])  # Eixo secundário (inset)

# Figura Principal
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Título Principal')

# Segunda Figura
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Título do eixo secundário');
````

---

#### 🔗 **`twinx()`: Plotando Gráficos com Escalas Diferentes**

O método **`twinx()`** do Matplotlib permite criar um **segundo eixo y** compartilhando o mesmo eixo x. Isso é útil quando precisamos **plotar duas curvas no mesmo gráfico**, mas com escalas verticais diferentes.

Ao usar **`twinx()`**, criamos um gráfico sobreposto ao original, mas com uma escala y independente, permitindo uma visualização clara e comparativa entre os conjuntos de dados.  

---

#### 📌 **Dica:**
Use **`twinx()`** quando precisar visualizar **duas variáveis relacionadas ao mesmo eixo x**, mas que apresentam escalas ou unidades distintas no eixo y.

````python
# Exemplo - Uso de duas escalas num mesmo plot - exemplo vindo do site do matplotlib
import numpy as np
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()
t = np.arange(0.01, 10.0, 0.01)
s1 = np.exp(t)
ax1.plot(t, s1, 'b-')
ax1.set_xlabel('time (s)')
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('exp', color='b')
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()
s2 = np.sin(2 * np.pi * t)
ax2.plot(t, s2, 'r.')
ax2.set_ylabel('sin', color='r')
ax2.tick_params('y', colors='r')

fig.tight_layout()
plt.show()
````

````python
# Exemplo definindo o tamanho dos pontos no método plt.scatter()

# Import numpy as np
import numpy as np

# Transformar a lista em Array: np_pop
np_pop = np.array(pop)

# Duplicar o valor de  np_pop
np_pop = np_pop * 2

# Atualização: configurar a variável s para np_pop
plt.scatter(gdp_cap, life_exp, s = np_pop)

# Customizações
plt.xscale('log')
plt.xlabel('PIB per Capita [em USD]')
plt.ylabel('Expectativa de Vida [em anos]')
plt.title('Desenvolvimento Global em 2007')
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])

# Mostar o plot
plt.show()
````

### Importante
Há muitos outros tipos de gráficos no Matplot
Por exemplo:
    - barras
    - errorbar
    - boxplot

````python
# Exemplo de uso de gráficos de errorbar

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x, y, yerr=dy, fmt='.k');
````

````python
# Exemplo de uso de gráficos de errorbar
# com barras em cinza claro

plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
ecolor='lightgray', elinewidth=3, capsize=0);
````

````python
# Exemplo de Gráfico de Pizza

import seaborn as sns
import matplotlib.pyplot as plt
txtLabels = 'Cats', 'Dogs', 'Frogs', 'Others'
fractions = [45, 30, 15, 10]
offsets =(0, 0.05, 0, 0)
plt.pie(fractions, explode=offsets, labels=txtLabels,
autopct='%1.1f%%', shadow=True, startangle=90,
colors=sns.color_palette('muted') )
plt.axis('equal')
````

````python
# Exemplos de gráficos boxplot - obtidos do site do Matplotlib

import matplotlib.pyplot as plt
import numpy as np

# Geração aleatória de alguns dados
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low), 0)

# Plot do boxplot básico
plt.boxplot(data)
````

````python
# Plot do boxplot entalhado
plt.figure()
plt.boxplot(data, 1);

plt.show()
````

````python
# Mudança de cor e formato dos pontos de outliers

plt.figure()
plt.boxplot(data, 0, 'gD');
````

````python
# Mudar o tamanho do bigode do boxplot

plt.boxplot(data, 0, 'rs', 0, whis=0.75);
````

````python
# Omitir os pontos de outliers
plt.figure()
plt.boxplot(data, 0, '');
````

````python
# Plot boxplot na horizontal
plt.figure()
plt.boxplot(data, vert=False);
````
